```python
df['is_family'] = ((df['number of children']) > 0) & (df['number of adults'] > 1)
df['is_family'] = df['is_family'].astype(int)  


# Counts
family_counts = df['is_family'].value_counts()



# Donut chart
wedges, texts, autotexts = plt.pie(
    family_counts,
    labels=["Non-Family", "Family"],
    autopct="%1.1f%%",                 # show only %
    startangle=90,
    textprops={'fontsize':14, 'weight':'bold'}
)

# Fix text styling
for autotext in autotexts:
    autotext.set_color("white")
    autotext.set_fontsize(14)
    autotext.set_weight("bold")


plt.tight_layout()
plt.show()


```

```python
df['date of reservation'] = pd.to_datetime(df['date of reservation'], errors='coerce')

df['year_month'] = df['date of reservation'].dt.to_period('M').astype(str)

monthly = df.groupby(['year_month', 'booking status']).size().unstack(fill_value=0)
monthly['Total'] = monthly.sum(axis=1)

if 'Canceled' in monthly.columns:
    monthly['CancelRate'] = monthly['Canceled'] 
else:
    monthly['CancelRate'] = 0
```

```python
plt.figure(figsize=(10,5))
monthly[['Total']].plot(kind='bar', legend=False, color='skyblue', alpha=0.7)
plt.title("Monthly Booking Counts")
plt.ylabel("Number of Bookings")
plt.xlabel("Year-Month")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
monthly['CancelRate'].plot(marker='o')
plt.title("Monthly Cancellation")
plt.ylabel("Amount Cancelled")
plt.xlabel("Year-Month")
plt.xticks(rotation=45)
plt.show()
```

```python
corr = dfEncoded.select_dtypes(include=['int64','float64']).corr()

plt.figure(figsize=(30, 20))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=1.0,
    linecolor='white',
    square=True,
    annot_kws={"size":12},  
)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.title("Correlation Matrix", pad=20, fontsize=16)

plt.tight_layout()
plt.show()
```
